```python
#Intermediate Python Concepts
## Recurrsion
# A Function calls itself


#Factorial of a number
#3! = 3*2*1
#4! = 4*3*2*1
#10! = 10*9*8*7*....1

#Used in Counting Theory Think of the example we saw in class - Selecting 5 board members from 200 people
#n!/(n-r)! r -> 5 (number of places)

#EDGE cases! -  0< n < 10000 !!!!!
0! = 1
```

```python
#write this without recursion

def factorial(num):
  fact = 1
  if (num==0) or (num==1):
    return 1
  else:
    for i in range(num,1,-1): #Every element in the iterable is being visited in this loop.
     fact*=i
  return fact
```

```python
factorial(2)
```

```python
#Recurssion is a type of a function taht calls itself
def fact_recurssion(n):
  if (n==1):
    return 1
  else:
    return n*fact_recurssion(n-1)
```
