#Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

#Step 2. Import the dataset from this address.

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv")
```

```python
df
```

#Step 3. Assign it to a variable called drinks.

```python
drinks= df
drinks.head()
```

#Step 4. Which continent drinks more beer on average?



```python
avg=df.groupby("continent").agg({"beer_servings":["mean"]})
```

```python
avg
```

#Step 5. For each continent print the statistics for wine consumption.

```python
st=df.groupby("continent").agg({"wine_servings": 'describe'})
```

```python
st
```

#Step 6. Print the mean alcoohol consumption per continent for every column

```python
st=df.groupby('continent').mean()
```

```python
st
```

#Step 7. Print the median alcoohol consumption per continent for every column

```python
st=df.groupby('continent').median()
```

```python
st
```
